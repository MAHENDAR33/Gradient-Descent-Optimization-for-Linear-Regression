```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Given Data
x1, x2 = 3, 2  # Inputs
y_actual = 17  # Expected Output

# Initialize Random Weights
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)
learning_rate = 0.01  # Step Size

# Store Error & Predictions for Visualization
errors = []
predictions = []

# Training Loop (20 Epochs)
for epoch in range(20):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2

    # Compute Squared Error
    error = (y_actual - y_pred) ** 2

    # Store Values
    errors.append(error)
    predictions.append(y_pred)

    # Compute Gradients (Backpropagation)
    dE_dw1 = -2 * (y_actual - y_pred) * x1
    dE_dw2 = -2 * (y_actual - y_pred) * x2

    # Update Weights
    w1 -= learning_rate * dE_dw1
    w2 -= learning_rate * dE_dw2

    # Print Progress
    print(f"Epoch {epoch+1}: y_pred = {y_pred:.4f}, Error = {error:.4f}")

# Visualization: Error Reduction Over Time
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(errors, label="Error", color='red')
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Error Reduction Over Training")
plt.legend()

# Visualization: Predictions Over Time
plt.subplot(1,2,2)
plt.plot(predictions, label="Predicted y", color='blue')
plt.axhline(y_actual, linestyle="--", color="green", label="Actual y")
plt.xlabel("Epochs")
plt.ylabel("Predicted y")
plt.title("Predicted y Over Training")
plt.legend()

plt.show()
```
